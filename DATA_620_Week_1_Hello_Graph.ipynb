```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
g = nx.Graph()

# The line of code below is not necessary
g.add_nodes_from(["Andre",
                  "Beverly",
                  "Ed",
                  "Carol",
                  "Diane",
                  "Fernando",
                  "Garth",
                  "Heather",
                  "Ike",
                  "Jane"])
pos = {"Andre":(-1,1),
                  "Beverly":(-1,-1),
                  "Carol":(0,2),
                  "Diane":(0,0),
                  "Ed":(0,-2),
                  "Fernando":(1,1),
                  "Garth":(1,-1),
                  "Heather":(2,0),
                  "Ike":(3,0),
                  "Jane":(4,0)}

g.add_edges_from([("Andre","Beverly"),
                  ("Andre","Carol"),
                  ("Andre","Diane"),
                  ("Andre","Fernando"),
                  ("Beverly","Diane"),
                  ("Beverly","Garth"),
                  ("Beverly","Ed"),
                  ("Carol","Fernando"),
                  ("Carol","Diane"),
                  ("Diane","Fernando"),
                  ("Diane","Ed"),
                  ("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Garth","Fernando"),
                  ("Heather", "Fernando"),
                  ("Heather","Garth"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])
nx.draw(g, pos, with_labels=True, node_size= 1000, font_size = 8, edge_color = "purple")
```

Diane has the highest number of connections with the other nodes. Ike and Jane will be disconnected from the graph if Heather is taken out.
